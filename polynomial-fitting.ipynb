```python
import numpy as np
import matplotlib.pyplot as plt

x = [0,1,2,3]
y = [3,2,8,10]
```

```python
%matplotlib inline
# small figure
plt.figure(figsize=(3,3))
plt.plot(x, y, 'x')
```

```python
# degree=0 function that is f(x) = 5.8
%matplotlib inline
# small figure
plt.figure(figsize=(3,3))
def f0(x):
    return 5.8

# and plot it
plt.plot(x, y, 'x')

# give x that is from 0 to 3
xs = np.linspace(0, 3, 100)
plt.plot(xs, [f0(x_) for x_ in xs])
```

```python
# degree=0 function that is f(x) = 5.8
%matplotlib inline
# small figure
plt.figure(figsize=(3,3))
def f1(x):
    return 3.05+1.85*x

# and plot it
plt.plot(x, y, 'x')

# give x that is from 0 to 3
xs = np.linspace(0, 3, 100)
plt.plot(xs, [f1(x_) for x_ in xs])
```
